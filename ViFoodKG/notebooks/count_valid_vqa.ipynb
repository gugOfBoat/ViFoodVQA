```python
from pathlib import Path
import os
from dotenv import load_dotenv
from supabase import create_client

PROJECT_ROOT = Path.cwd().parent
load_dotenv(PROJECT_ROOT / ".env")

SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_KEY")
supabase = create_client(SUPABASE_URL, SUPABASE_KEY)
```

### Đếm số VQA chưa check trong split test

```python
start_vqa_id = 1
end_vqa_id = 14381
target_split = "test"

resp = (
    supabase.table("vqa")
    .select("vqa_id", count="exact")
    .gte("vqa_id", start_vqa_id)
    .lte("vqa_id", end_vqa_id)
    .eq("split", target_split)
    .eq("is_checked", False)
    .eq("is_drop", False)
    .execute()
)

unchecked_count = resp.count if resp.count is not None else 0

print(f"Số câu VQA chưa được check trong split='{target_split}', range [{start_vqa_id}, {end_vqa_id}]: {unchecked_count}")
```

### Đếm số VQA hợp lệ trong split test

```python
start_vqa_id = 1
end_vqa_id = 14381
target_split = "test"

resp = (
    supabase.table("vqa")
    .select("vqa_id", count="exact")
    .gte("vqa_id", start_vqa_id)
    .lte("vqa_id", end_vqa_id)
    .eq("split", target_split)
    .eq("is_checked", True)
    .eq("is_drop", False)
    .execute()
)

valid_count = resp.count if resp.count is not None else 0

print(f"Số câu VQA hợp lệ trong split='{target_split}', range [{start_vqa_id}, {end_vqa_id}]: {valid_count}")
```
